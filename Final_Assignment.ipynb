```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
#import csv's from sports reference NBA
nba_2020_season = pd.read_csv("2020_nba_advanced.csv", index_col = 'Team')
nba_2020_season
nba_2021_season = pd.read_csv("2021_nba_advanced.csv", index_col = 'Team')
nba_2021_season
nba_2022_season = pd.read_csv("2022_nba_advanced.csv", index_col = 'Team')
nba_2022_season
```

```python
nba_season_2020_2022 = pd.concat([nba_2020_season,nba_2021_season,nba_2022_season], ignore_index=True)
nba_season_2020_2022
```

```python
nba_season_2020_2022['Win_Percentage'] = (nba_season_2020_2022['W'] / (nba_season_2020_2022['W'] + nba_season_2020_2022['L']) * 100)
nba_season_2020_2022
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(nba_season_2020_2022.Win_Percentage)
ax2.scatter(nba_season_2020_2022.Win_Percentage,nba_season_2020_2022.Pace)
ax3.scatter(nba_season_2020_2022.Win_Percentage,nba_season_2020_2022.NRtg)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title("NBA Win Percentage")
ax.set_xlabel("Win Percentage")
ax.set_ylabel("Number of Teams")
ax.hist(nba_season_2020_2022.Win_Percentage)
fig.savefig("nba_win_pct.pdf")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title("NBA Win Percentage with Net Rating")
ax.set_xlabel("Win Percentage")
ax.set_ylabel("Net Rating")
ax.scatter(nba_season_2020_2022.Win_Percentage,nba_season_2020_2022.NRtg)
fig.savefig("nba_win_pct_nrtg_scatter.pdf")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title("NBA Win Percentage with Pace of Play")
ax.set_xlabel("Win Percentage")
ax.set_ylabel("Pace of Play")
ax.scatter(nba_season_2020_2022.Win_Percentage,nba_season_2020_2022.Pace)
fig.savefig("nba_win_pct_pace.pdf")
```

```python
nba_selected_data = nba_season_2020_2022[['W','L','Pace','NRtg', 'Off_eFG%']]
nba_selected_data
```

```python
nba_selected_data = nba_season_2020_2022[['W','L','Pace','NRtg', 'Off_eFG%']]
sns.pairplot(nba_selected_data, diag_kind="kde", kind="reg", 
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws':{'alpha':0.2}}).savefig("scatter_plot_matrix_nba")
```

```python
win_loss_percentage_selected_data = nba_season_2020_2022[['Win_Percentage','NRtg','Pace']]
win_loss_percentage_selected_data
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax1.set_xlabel("Net Rating")
ax1.set_ylabel("Number of Teams")
ax1.scatter(win_loss_percentage_selected_data.NRtg,win_loss_percentage_selected_data.Win_Percentage)
ax2.set_xlabel("Net Rating")
ax2.set_ylabel("Frequency")
ax2.hist(win_loss_percentage_selected_data.NRtg)
sns.regplot(x=win_loss_percentage_selected_data.NRtg, y=win_loss_percentage_selected_data.Win_Percentage, ax=ax1)
fig.savefig("nba_win_pct_net_rating.pdf")
```
