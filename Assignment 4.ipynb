```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
```

```python
knicks_roster = pd.read_csv("knicks_2122_roster.csv", index_col = 'Player')
knicks_roster
lakers_roster = pd.read_csv("lakers_2122_roster.csv", index_col = 'Player')
lakers_roster
nets_roster = pd.read_csv("nets_2122_roster.csv", index_col = 'Player')
nets_roster
warriors_roster = pd.read_csv("warriors_2122_roster.csv", index_col = 'Player')
warriors_roster
```

```python
#Combine dataframes into single dataframe table
nba_2122_rosters = pd.concat([knicks_roster,lakers_roster,nets_roster,warriors_roster])
nba_2122_rosters
```

```python
nba_2122_rosters.to_csv("NBA_2122_rosters_combined")
```
